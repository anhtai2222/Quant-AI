# 🚀 PROJECT: THE XGBOOST DOW-THEORY HUNTER (ETH/USDT M30)

**Mục tiêu chiến lược:**
- **Phong cách:** Trend Following (Thuận xu hướng) + Price Action + ML Filter.
- **Khung thời gian:** M30.
- **Hiệu suất kỳ vọng:** Sharpe > 0.3, Annualized Return > 15%, Vượt Buy & Hold.

**Quy trình xử lý:**
1. **Clean Data:** Loại bỏ nhiễu, volume ảo.
2. **Feature Engineering:** Biến đổi nến thành các chỉ số Toán học (Dow, RVOL, ATR).
3. **Train AI:** Dạy XGBoost nhận diện setup thắng.
4. **Backtest:** Giả lập giao dịch với quy tắc thoát lệnh chặt chẽ (Time-stop, Dynamic R:R).

```python
# 1. IMPORT THƯ VIỆN & CẤU HÌNH
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, precision_score
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

# Cấu hình đồ thị
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. DATA LOADER & CLEANING (LÀM SẠCH DỮ LIỆU)
Bước này cực kỳ quan trọng. Dữ liệu Crypto thường chứa các cây nến có Volume = 0 (do mất kết nối hoặc bảo trì sàn). Nếu không lọc, các chỉ báo như RVOL sẽ bị sai lệch.

```python
def load_and_clean_data(filepath):
    print("--- ĐANG TẢI VÀ LÀM SẠCH DỮ LIỆU ---")
    # Đọc file CSV
    df = pd.read_csv(filepath)
    
    # Chuẩn hóa tên cột về chữ thường
    df.columns = [col.lower().strip() for col in df.columns]
    
    # Chuyển đổi timestamp
    if 'timestamp' in df.columns:
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df.set_index('timestamp', inplace=True)
    
    # Sắp xếp thời gian
    df.sort_index(inplace=True)
    
    # --- CLEANING ---
    initial_len = len(df)
    
    # 1. Loại bỏ duplicate index
    df = df[~df.index.duplicated(keep='first')]
    
    # 2. Loại bỏ các nến Volume = 0 (Thị trường chết/Sàn lỗi)
    df = df[df['volume'] > 0]
    
    # 3. Loại bỏ các nến bị lỗi dữ liệu (High < Low)
    df = df[df['high'] >= df['low']]
    
    # 4. Fill các giá trị NaN nhỏ (nếu có)
    df = df.ffill()
    
    print(f"Dữ liệu gốc: {initial_len} dòng")
    print(f"Dữ liệu sạch: {len(df)} dòng (Đã loại bỏ {initial_len - len(df)} dòng nhiễu)")
    
    return df

# LOAD DATA (Thay đổi đường dẫn nếu cần)
file_path = 'ETHUSDT.csv' 
df = load_and_clean_data(file_path)

# Xem qua dữ liệu
df.tail()
```

## 3. FEATURE ENGINEERING (SỐ HÓA PRICE ACTION)
Chúng ta sẽ không dùng thư viện ngoài (`ta`) để code dễ chạy ở mọi môi trường. Các chỉ báo sẽ được tính thủ công bằng Pandas.

**Các đặc trưng chính:**
1. **Trend Filter:** `EMA_400` (Tương đương EMA 50 khung H4).
2. **Dow Patterns:** `Higher High`, `Higher Low`, `Breakout`.
3. **Price Action:** `Lower Shadow` (Pinbar), `Body Strength`.
4. **Volume:** `RVOL` (Relative Volume).

```python
def create_features(df):
    data = df.copy()
    
    # --- A. TREND INDICATORS ---
    # EMA 400 trên M30 ~ EMA 50 trên H4 (Trend trung hạn)
    data['ema_trend'] = data['close'].ewm(span=400, adjust=False).mean()
    data['dist_ema'] = (data['close'] - data['ema_trend']) / data['close']
    
    # --- B. VOLATILITY (ATR) ---
    # Tính True Range
    data['tr1'] = data['high'] - data['low']
    data['tr2'] = abs(data['high'] - data['close'].shift(1))
    data['tr3'] = abs(data['low'] - data['close'].shift(1))
    data['tr'] = data[['tr1', 'tr2', 'tr3']].max(axis=1)
    # ATR 14
    data['atr'] = data['tr'].rolling(14).mean()
    data['atr_pct'] = data['atr'] / data['close']
    
    # --- C. VOLUME ANOMALY (RVOL) ---
    # Volume hiện tại so với trung bình 20 cây nến trước
    data['vol_ma20'] = data['volume'].rolling(20).mean()
    data['rvol'] = data['volume'] / (data['vol_ma20'] + 1e-9)
    
    # --- D. DOW THEORY STRUCTURE ---
    # Đỉnh sau cao hơn đỉnh trước
    data['is_HH'] = np.where(data['high'] > data['high'].shift(1), 1, 0)
    # Đáy sau cao hơn đáy trước
    data['is_HL'] = np.where(data['low'] > data['low'].shift(1), 1, 0)
    
    # Breakout 20 nến (Phá nền giá)
    data['recent_high'] = data['high'].rolling(20).max().shift(1)
    data['is_breakout'] = np.where(data['close'] > data['recent_high'], 1, 0)
    
    # --- E. PRICE ACTION (NẾN) ---
    # Lower Shadow (Râu dưới - Thể hiện lực bắt đáy)
    # Công thức: (Min(Open, Close) - Low) / Range
    body_bottom = np.minimum(data['open'], data['close'])
    data['candle_range'] = data['high'] - data['low']
    data['lower_shadow'] = (body_bottom - data['low']) / (data['candle_range'] + 1e-9)
    
    # Body Relative Size (Thân nến dài hay ngắn)
    data['body_size'] = abs(data['close'] - data['open']) / (data['candle_range'] + 1e-9)
    
    # Momentum Returns
    data['ret_1'] = data['close'].pct_change(1)
    
    # --- F. LABELING (MỤC TIÊU CHO AI HỌC) ---
    # Mục tiêu: Giá tăng ít nhất 2 ATR trong vòng 4 nến tới (2 giờ)
    look_ahead = 4
    profit_threshold_atr = 2.0
    
    future_high = data['high'].shift(-look_ahead).rolling(look_ahead).max()
    
    # Nhãn 1 nếu (Future High - Close) > 2 * ATR
    data['target'] = np.where(
        (future_high - data['close']) > (data['atr'] * profit_threshold_atr), 
        1, 0
    )
    
    # Drop NaN do rolling
    return data.dropna()

# Áp dụng
df_processed = create_features(df)
print("Đã tạo xong Features. Số lượng mẫu dùng để train:", len(df_processed))
```

## 4. TRAINING XGBOOST MODEL
Chúng ta chia dữ liệu theo thời gian (Time-series split): 80% quá khứ để học, 20% gần nhất để backtest. Tuyệt đối không được xáo trộn (Shuffle) dữ liệu.

```python
def train_model(data):
    # Chọn Features đầu vào cho AI
    features = [
        'rvol', 'atr_pct', 'lower_shadow', 'body_size', 
        'is_HH', 'is_HL', 'is_breakout', 'dist_ema', 'ret_1'
    ]
    
    X = data[features]
    y = data['target']
    
    # Chia Train/Test (80% / 20%)
    split_point = int(len(X) * 0.8)
    X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]
    y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]
    
    # Cấu hình XGBoost (Tối ưu để tránh Overfitting)
    model = xgb.XGBClassifier(
        n_estimators=150,       # Số cây
        learning_rate=0.03,     # Học chậm
        max_depth=5,            # Độ sâu cây
        subsample=0.8,          # Chọn ngẫu nhiên mẫu
        colsample_bytree=0.8,   # Chọn ngẫu nhiên feature
        random_state=42,
        n_jobs=-1,
        eval_metric='logloss'
    )
    
    print("Bot đang học...")
    model.fit(X_train, y_train)
    
    # Đánh giá trên tập Test
    y_pred = model.predict(X_test)
    print("\n--- KẾT QUẢ ĐÁNH GIÁ TRÊN TẬP TEST ---")
    print(classification_report(y_test, y_pred))
    
    # Feature Importance
    xgb.plot_importance(model, max_num_features=10, height=0.5)
    plt.title("Các yếu tố AI coi trọng nhất")
    plt.show()
    
    return model, features, split_point

model, features, split_index = train_model(df_processed)
```

## 5. BACKTEST ENGINE (MÔ PHỎNG THỰC TẾ)
Đây là phần quan trọng nhất để xác định Sharpe Ratio và ROI.
Chúng ta áp dụng bộ quy tắc:
1. **Entry:** `AI Prob > 0.60` VÀ `Price > EMA 400`.
2. **Stop Loss:** `Entry - 2 * ATR`.
3. **Take Profit:** `Entry + 5 * ATR` (R:R = 1:2.5).
4. **Time Stop:** Đóng lệnh sau 6 nến (3 tiếng) nếu chưa chạm TP/SL.

```python
def run_backtest(data, model, features, split_idx):
    print("\n--- BẮT ĐẦU BACKTEST (TRÊN TẬP TEST) ---")
    
    # Chỉ lấy dữ liệu test (Tương lai giả định)
    df_test = data.iloc[split_idx:].copy()
    
    # 1. AI Dự đoán xác suất
    df_test['ai_prob'] = model.predict_proba(df_test[features])[:, 1]
    
    # 2. Logic Vào lệnh (Trend Filter + AI)
    # Ngưỡng AI: 0.60 (60% tự tin)
    df_test['entry_signal'] = np.where(
        (df_test['ai_prob'] > 0.60) & (df_test['close'] > df_test['ema_trend']), 
        1, 0
    )
    
    # 3. Vòng lặp mô phỏng PnL
    capital = 10000.0 # Vốn khởi đầu $10,000
    balance = [capital]
    trades_pnl = []
    holding = False
    entry_idx = 0
    
    # Chuẩn bị dữ liệu Buy & Hold
    start_price = df_test.iloc[0]['close']
    end_price = df_test.iloc[-1]['close']
    bh_return = (end_price - start_price) / start_price
    
    for i in range(len(df_test)):
        # Nếu đang không giữ lệnh và có tín hiệu Mua
        if not holding and df_test.iloc[i]['entry_signal'] == 1:
            entry_price = df_test.iloc[i]['close']
            atr = df_test.iloc[i]['atr']
            
            # Setup TP/SL
            sl_price = entry_price - (2.0 * atr)
            tp_price = entry_price + (5.0 * atr) # R:R = 1:2.5
            
            holding = True
            entry_idx = i
            entry_time = df_test.index[i]
            continue
            
        # Nếu đang giữ lệnh, kiểm tra điều kiện thoát
        if holding:
            curr_low = df_test.iloc[i]['low']
            curr_high = df_test.iloc[i]['high']
            curr_close = df_test.iloc[i]['close']
            
            pnl = 0
            exit_reason = ""
            
            # 1. Check SL
            if curr_low <= sl_price:
                pnl = -0.02 # Cắt lỗ 2% tài khoản (Fixed Risk)
                exit_reason = "SL"
            
            # 2. Check TP
            elif curr_high >= tp_price:
                # Risk 2%, Reward 5% (do R:R = 1:2.5)
                pnl = 0.05 
                exit_reason = "TP"
            
            # 3. Check Time Stop (6 nến = 3 tiếng)
            elif (i - entry_idx) >= 6:
                # Thoát lệnh ở giá đóng cửa
                raw_return = (curr_close - entry_price) / entry_price
                # Quy đổi risk: SL 2% ~ khoảng cách 2 ATR
                # PnL thực tế = Risk_Per_Trade * (Raw_Return / SL_Distance_Pct)
                sl_dist_pct = (entry_price - sl_price) / entry_price
                pnl = 0.02 * (raw_return / sl_dist_pct)
                exit_reason = "TIME"
            
            # Nếu có lý do thoát -> Đóng lệnh
            if exit_reason != "":
                # Cập nhật Balance (Lãi kép)
                new_balance = balance[-1] * (1 + pnl)
                balance.append(new_balance)
                trades_pnl.append(pnl)
                holding = False
                
    # --- TÍNH TOÁN METRICS ---
    final_balance = balance[-1]
    total_return = (final_balance - capital) / capital
    
    # Annualized Return (Giả sử test trên khoảng thời gian X năm)
    days_test = (df_test.index[-1] - df_test.index[0]).days
    annualized_return = ((1 + total_return) ** (365 / days_test)) - 1
    
    # Sharpe Ratio (Ước lượng)
    if len(trades_pnl) > 0:
        returns_series = pd.Series(trades_pnl)
        # Sharpe = (Mean Return / Std Dev) * sqrt(Số lệnh trung bình năm)
        # Đây là Sharpe của Trade PnL
        avg_trades_per_year = len(trades_pnl) * (365 / days_test)
        sharpe = (returns_series.mean() / returns_series.std()) * np.sqrt(avg_trades_per_year)
        win_rate = len(returns_series[returns_series > 0]) / len(returns_series)
    else:
        sharpe = 0; win_rate = 0
    
    # --- REPORT ---
    print(f"\n=== KẾT QUẢ HIỆU SUẤT ({days_test} NGÀY) ===")
    print(f"Vốn ban đầu:      ${capital:,.2f}")
    print(f"Vốn cuối cùng:    ${final_balance:,.2f}")
    print(f"Tổng số lệnh:     {len(trades_pnl)}")
    print(f"Tỷ lệ thắng:      {win_rate:.2%}")
    print(f"-----------------------------")
    print(f"Chiến lược Return:{total_return:.2%}")
    print(f"Buy & Hold Return:{bh_return:.2%}")
    print(f"-----------------------------")
    print(f"CAGR (Annualized):{annualized_return:.2%}  (Mục tiêu > 15%)")
    print(f"Sharpe Ratio:     {sharpe:.2f}       (Mục tiêu > 0.3)")
    
    # Plot Equity Curve
    plt.figure(figsize=(12, 6))
    plt.plot(balance, label='AI Strategy Equity')
    plt.title('Đường cong tăng trưởng tài khoản')
    plt.xlabel('Số lệnh')
    plt.ylabel('Vốn ($)')
    plt.legend()
    plt.show()
    
    return balance

equity = run_backtest(df_processed, model, features, split_index)
```
