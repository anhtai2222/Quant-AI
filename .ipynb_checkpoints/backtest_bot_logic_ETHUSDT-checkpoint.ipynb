# Backtest Bot 2 tầng (Entry/Exit) + RR Gate + ATR Volatility Scaling (ETHUSDT)

**Dataset:** `ETHUSDT.csv`  
**Timeframe ước tính:** ~30 phút/bar | **Annualization:** 17520 bars/năm

Notebook này:
- Hiển thị rõ **logic vào/ra lệnh** theo mô tả của bạn (Bot 1 Entry, Bot 2 RR gate, Trend filter EMA200, ATR-based dynamic SL/TP).
- Xuất **thống kê backtest**: Sharpe, Max Drawdown, Annual Return (CAGR), Win rate, Profit factor, Trades/year, v.v.
- Có cell **grid tuning** (rr_ratio, look_ahead, risk_per_trade).

> Mặc định backtest theo giá OHLC, **không** mô phỏng funding. Có tham số fee/slippage (bps) để bạn bật/tắt.


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

CSV_PATH = r"/mnt/data/ETHUSDT.csv"

def load_ohlc(path: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    df.columns = [c.lower() for c in df.columns]
    ts_col = "timestamp" if "timestamp" in df.columns else ("date" if "date" in df.columns else df.columns[0])
    df[ts_col] = pd.to_datetime(df[ts_col])
    df = df.sort_values(ts_col).set_index(ts_col)
    required = ["open","high","low","close"]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing columns: {missing}. Found: {df.columns.tolist()}")
    return df

df = load_ohlc(CSV_PATH)
df.head()
```

## 1) Indicators: EMA200 + ATR
- EMA200: Trend filter
- ATR(14): đo biến động để scale SL/TP


```

def ema(series: pd.Series, span: int) -> pd.Series:
    return series.ewm(span=span, adjust=False).mean()

def atr(df: pd.DataFrame, period: int = 14) -> pd.Series:
    high, low, close = df["high"], df["low"], df["close"]
    prev_close = close.shift(1)
    tr = pd.concat([
        (high - low),
        (high - prev_close).abs(),
        (low - prev_close).abs()
    ], axis=1).max(axis=1)
    return tr.ewm(span=period, adjust=False).mean()

df["ema200"] = ema(df["close"], 200)
df["atr"] = atr(df, 14)

df[["close","ema200","atr"]].dropna().tail()
```

## 2) Params (tune theo mục tiêu Sharpe/ROI)
- `rr_ratio`: Bot 2 hủy lệnh nếu TP/SL < rr_ratio  
- `look_ahead`: horizon dự báo của regression  
- `risk_per_trade`: % equity rủi ro cho mỗi lệnh (size tự động scale theo SL distance)  
- `sl_atr_mult`, `tp_atr_mult`: SL/TP theo ATR  
- `fee_bps`, `slippage_bps`: bật/tắt mô phỏng chi phí


```

from dataclasses import dataclass

@dataclass
class Params:
    look_ahead: int = 12
    reg_window: int = 96
    rr_ratio: float = 1.5
    sl_atr_mult: float = 2.0
    tp_atr_mult: float = 3.0
    risk_per_trade: float = 0.02
    fee_bps: float = 2.0
    slippage_bps: float = 1.0
    allow_reentry_same_bar: bool = False

P = Params()
P
```

## 3) Bot 1 (Entry): Regression forecast
Hồi quy tuyến tính trên **log(close)** trong `reg_window` bars rồi extrapolate tới `look_ahead`.

- Bot 1 "muốn Long" nếu `forecast_price > close`
- Bot 2 dùng `expected_move = |forecast_price - close|` để lọc quality (đủ lớn để ăn TP).


```

def regression_forecast_price(close: pd.Series, reg_window: int, look_ahead: int) -> pd.Series:
    y = np.log(close.values)
    n = len(y)
    out = np.full(n, np.nan)

    x = np.arange(reg_window)
    x_mean = x.mean()
    x_demean = x - x_mean
    denom = (x_demean**2).sum()

    for i in range(reg_window, n - look_ahead):
        window = y[i-reg_window:i]
        y_mean = window.mean()
        y_demean = window - y_mean
        beta = (x_demean * y_demean).sum() / denom
        alpha = y_mean - beta * x_mean
        x_pred = (reg_window - 1) + look_ahead
        y_pred = alpha + beta * x_pred
        out[i] = np.exp(y_pred)

    return pd.Series(out, index=close.index, name="forecast_price")

df["forecast_price"] = regression_forecast_price(df["close"], P.reg_window, P.look_ahead)
df["exp_move"] = (df["forecast_price"] - df["close"]).abs()
df["exp_move_atr"] = df["exp_move"] / df["atr"]

df[["close","forecast_price","atr","exp_move_atr"]].dropna().tail()
```

## 4) Backtest engine (rõ logic vào/ra)
### Điều kiện vào lệnh (Long-only)
1. Trend filter: `close > EMA200`
2. Bot 1: `forecast_price > close`
3. Bot 2 RR gate:
   - SL = `sl_atr_mult * ATR`
   - TP = `tp_atr_mult * ATR`
   - RR = TP/SL >= `rr_ratio`
   - Và `expected_move >= TP` (nếu forecast không đủ lớn để ăn TP thì hủy)

### Quản trị rủi ro (Risk-per-trade cố định)
- `risk_amount = equity * risk_per_trade`
- `size = risk_amount / SL_distance`

### Thoát lệnh
- Nếu trong 1 bar:
  - `low <= SL` → Stop
  - `high >= TP` → Take profit
- Nếu hit cả TP và SL trong cùng bar: mặc định **SL-first** (conservative), có thể đổi.


```

from dataclasses import dataclass

@dataclass
class Trade:
    entry_time: pd.Timestamp
    exit_time: pd.Timestamp
    entry_price: float
    exit_price: float
    size: float
    pnl: float
    reason: str

def backtest(df: pd.DataFrame, P: Params, tp_first: bool = False):
    data = df.dropna(subset=["ema200","atr","forecast_price"]).copy()
    idx = data.index

    equity = 1.0
    equity_curve = []
    times = []

    in_pos = False
    entry_price = sl_price = tp_price = size = 0.0
    entry_time = None
    trades = []

    fee_rt = P.fee_bps / 1e4        # round-trip
    slip_side = P.slippage_bps / 1e4

    def px_entry(px): return px * (1 + slip_side)
    def px_exit(px):  return px * (1 - slip_side)

    for i in range(1, len(data)):
        t = idx[i]
        prev = data.iloc[i-1]
        row = data.iloc[i]
        o, h, l = row["open"], row["high"], row["low"]

        # 1) manage position: check bracket hits inside this bar
        if in_pos:
            hit_tp = h >= tp_price
            hit_sl = l <= sl_price
            exit_px = None
            reason = None

            if hit_tp and hit_sl:
                if tp_first:
                    exit_px, reason = tp_price, "TP&SL (TP first)"
                else:
                    exit_px, reason = sl_price, "TP&SL (SL first)"
            elif hit_tp:
                exit_px, reason = tp_price, "TP"
            elif hit_sl:
                exit_px, reason = sl_price, "SL"

            if reason is not None:
                ex = px_exit(exit_px)
                gross_pnl = (ex - entry_price) * size
                fee = fee_rt * entry_price * size
                pnl = gross_pnl - fee

                # equity update (equity is in "units")
                equity = equity * (1 + pnl / equity)

                trades.append(Trade(entry_time, t, entry_price, ex, size, pnl, reason))
                in_pos = False

        # 2) entry logic: use prev bar signal, execute at current open
        if (not in_pos) or P.allow_reentry_same_bar:
            uptrend = prev["close"] > prev["ema200"]
            if uptrend:
                wants_long = prev["forecast_price"] > prev["close"]
                if wants_long:
                    atr_val = prev["atr"]
                    if atr_val > 0 and np.isfinite(atr_val):
                        sl_dist = P.sl_atr_mult * atr_val
                        tp_dist = P.tp_atr_mult * atr_val
                        rr = tp_dist / sl_dist if sl_dist > 0 else 0.0

                        supports_tp = prev["exp_move"] >= tp_dist

                        if rr >= P.rr_ratio and supports_tp:
                            risk_amount = equity * P.risk_per_trade
                            size = risk_amount / sl_dist
                            en = px_entry(o)

                            entry_price = en
                            entry_time = t
                            sl_price = en - sl_dist
                            tp_price = en + tp_dist
                            in_pos = True

        times.append(t)
        equity_curve.append(equity)

    equity_series = pd.Series(equity_curve, index=pd.DatetimeIndex(times), name="equity")
    return equity_series, trades

equity, trades = backtest(df, P, tp_first=False)
len(trades), equity.iloc[-1]
```

## 5) Backtest stats (Sharpe, Drawdown, Annual Return, Win rate, v.v.)

```

def infer_bars_per_year(index: pd.DatetimeIndex) -> int:
    diffs = index.to_series().diff().dropna()
    if len(diffs) == 0:
        return 365
    med = diffs.median()
    minutes = max(1, int(round(med.total_seconds()/60)))
    bars_per_day = int(round(24*60/minutes))
    return bars_per_day * 365

def sharpe_annualized(equity: pd.Series, bars_per_year: int) -> float:
    r = equity.pct_change().dropna()
    if r.std() == 0 or len(r) < 2:
        return float("nan")
    return float((r.mean()/r.std()) * np.sqrt(bars_per_year))

def max_drawdown(equity: pd.Series) -> float:
    dd = equity / equity.cummax() - 1.0
    return float(dd.min())

def cagr(equity: pd.Series) -> float:
    if len(equity) < 2:
        return float("nan")
    start, end = equity.index[0], equity.index[-1]
    years = (end - start).total_seconds() / (365*24*3600)
    if years <= 0:
        return float("nan")
    return float((equity.iloc[-1] / equity.iloc[0]) ** (1/years) - 1)

def trade_metrics(trades):
    if not trades:
        return {
            "Trades": 0, "Win rate": 0.0, "Avg win": 0.0, "Avg loss": 0.0,
            "Profit factor": float("nan"), "Expectancy": 0.0
        }
    pnls = np.array([t.pnl for t in trades], dtype=float)
    wins = pnls[pnls > 0]
    losses = pnls[pnls < 0]
    win_rate = float((pnls > 0).mean())
    avg_win = float(wins.mean()) if len(wins) else 0.0
    avg_loss = float(losses.mean()) if len(losses) else 0.0
    profit_factor = float(wins.sum() / (-losses.sum())) if len(losses) else float("inf")
    expectancy = float(pnls.mean())
    return {
        "Trades": int(len(trades)),
        "Win rate": win_rate,
        "Avg win": avg_win,
        "Avg loss": avg_loss,
        "Profit factor": profit_factor,
        "Expectancy (avg pnl/trade)": expectancy,
    }

bpy = infer_bars_per_year(equity.index)
years = (equity.index[-1] - equity.index[0]).total_seconds() / (365*24*3600)

summary = {
    "Period (years)": float(years),
    "Final Equity (x)": float(equity.iloc[-1]),
    "ROI": float(equity.iloc[-1]/equity.iloc[0] - 1),
    "Annual Return (CAGR)": cagr(equity),
    "Sharpe (ann.)": sharpe_annualized(equity, bpy),
    "Max Drawdown": max_drawdown(equity),
}
summary.update(trade_metrics(trades))

# Trades/year
summary["Trades/year"] = float(summary["Trades"] / years) if years > 0 else float("nan")

pd.Series(summary).to_frame("value")
```

## 8) Quick tuning grid (rr_ratio, look_ahead, risk)
Xem top configs theo Sharpe/ROI.


```

def run_once(params: Params):
    # recompute forecast if look_ahead changed
    tmp = df.copy()
    tmp["forecast_price"] = regression_forecast_price(tmp["close"], params.reg_window, params.look_ahead)
    tmp["exp_move"] = (tmp["forecast_price"] - tmp["close"]).abs()
    eq, tr = backtest(tmp, params, tp_first=False)
    bpy = infer_bars_per_year(eq.index)
    yrs = (eq.index[-1] - eq.index[0]).total_seconds() / (365*24*3600)
    return {
        "look_ahead": params.look_ahead,
        "rr_ratio": params.rr_ratio,
        "risk": params.risk_per_trade,
        "sl_atr": params.sl_atr_mult,
        "tp_atr": params.tp_atr_mult,
        "ROI": float(eq.iloc[-1]/eq.iloc[0]-1),
        "CAGR": cagr(eq),
        "Sharpe": sharpe_annualized(eq, bpy),
        "MaxDD": max_drawdown(eq),
        "Trades": len(tr),
        "WinRate": float(np.mean([t.pnl>0 for t in tr])) if tr else 0.0,
        "Trades/year": float(len(tr)/yrs) if yrs>0 else np.nan,
    }

grid = []
for look_ahead in [12, 16, 24]:
    for rr_ratio in [1.5, 2.0]:
        for risk in [0.02, 0.03, 0.04]:
            params = Params(
                look_ahead=look_ahead,
                reg_window=P.reg_window,
                rr_ratio=rr_ratio,
                sl_atr_mult=P.sl_atr_mult,
                tp_atr_mult=P.tp_atr_mult,
                risk_per_trade=risk,
                fee_bps=P.fee_bps,
                slippage_bps=P.slippage_bps,
            )
            try:
                grid.append(run_once(params))
            except Exception as e:
                grid.append({"look_ahead": look_ahead, "rr_ratio": rr_ratio, "risk": risk, "error": str(e)})

res = pd.DataFrame(grid)
res.sort_values(["Sharpe","ROI"], ascending=False).head(20)
```
